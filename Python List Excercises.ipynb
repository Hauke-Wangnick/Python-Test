## 1.
Write a Python program to sum all the items in a list 

```python
l = [1,5,22,1,666,1,2,0,91]

summe = 0
for i in l:
    summe += i
print(summe)
```

## 3.
Write a Python program to get the largest number from a list 


```python
l = [1,5,22,1,666,1,2,91]
#l = [1,2,3,4]

l1 = sorted(l,reverse=True)
print(f'Variante 1 ergibt: {l1[0]} als größten Wert der Liste')

l2 = sorted(l)
print(f'Variante 2 ergibt: {l2.pop()} als größten Wert der Liste')


result = l[0]
for i in l:
    if i > result:
        result = i
print(f'Variante 3 ergibt: {result} als größten Wert der Liste')
```

## 5.
Write a Python program to count the number of strings where the string length is 2 or more and the first and last character are same. The given list of strings could be: ['abc', 'xyz', 'aba', '1221']Expected Result : 2




```python
l = ['abc','xyz','aba','1221','125151','11','aa','a','1'] #Erwartet 5:'aba','1221','1255151','11','aa'

counter = 0
for i in l:
    if len(i) >= 2 and i[0] == i[-1]:
        counter += 1

print(counter)
```

```python
def common_member(l1,l2):
    for el in l1:
        if el in l2:
            return True       
    return False
```

```python
def add_y_behind_x(l,x,y):
    for index,el in enumerate(l):
        if type(el) == list:
            add_y_behind_x(el,x,y)
        elif el == x:
            l.insert(index+1,y)
    return(l)
```

```python
l1 = [10, 20, [300, 400, [5000, 6000], 500], 30, 40]
l2 = [10, 20, [300, 6000, 400, [5000, 6000], 500], 30, 40, 6000]
x = 6000
y = 7000

l = l2
print(l)
print(add_y_behind_x(l,x,y))
```
